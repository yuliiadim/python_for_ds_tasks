````python
from math import pi

def calc_area(r):                      #Використала стандарт написання функцій з маленької літери та з підкресленнями
    """Обчислює площу кола."""
    return pi * r ** 2
                                    #порожні рядкі між написанням функцій та пробіли навколо бінарних операторів
                                    #Додала рядки документації (docstrings) для пояснення призначення функцій.
                                    #Виключила використання * у імпорті, замість цього імпортувала лише pi.
def calc_circumference(r):
    """Обчислює circumference кола."""
    return 2 * pi * r


radius = 5
print("Area:", calc_area(radius), "Circumference:", calc_circumference(radius))
````

````python
def calculate_variance(numbers):
    """Calculate the variance of a list of numbers.

    Args:
        numbers (list of float): A list of numerical values.

    Returns:
        float: The variance of the numbers in the list.

    Raises:
        ValueError: If the input list is empty.
    """
    if not numbers:
        raise ValueError("The list cannot be empty")

    mean = sum(numbers) / len(numbers)
    mean_squares = sum(n ** 2 for n in numbers) / len(numbers)
    variance = mean_squares - mean ** 2
    
    return variance
````
